```python
#Mengubah File xlsx Menjadi csv
read_file = pd.read_excel ("DatasetCuaca_convert.xlsx")
read_file.to_csv ("DatasetCuaca_convert.csv", index = None, header=True)
df = pd.DataFrame(pd.read_csv("DatasetCuaca_convert.csv"))

#Menghilangkan Variabel yang tidak digunakan
df.drop('Bulan dan Tahun', axis=1, inplace=True)
df
```

```python
#Mencari Nilai Mean, Median dan Modus dari Suhu Udara
mean = df['Suhu_Udara'].mean()
median = df['Suhu_Udara'].median()
modus = df['Suhu_Udara'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus =', modus)
```

```python
#Mencari Nilai Mean, Median dan Modus dari Curah Hujan
mean = df['Curah_Hujan'].mean()
median = df['Curah_Hujan'].median()
modus = df['Curah_Hujan'].mode()[0]
print('mean =', mean)
print('median =', median)
print('modus =', modus)
```

```python
#visualisasi Data Kelembaban Udara dan Curah Hujan

#Visualisasi Dengan Boxplot
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.boxplot(df['Kelembapan_Udara'])
plt.title('Boxplot Kelembapan udara')

plt.subplot(1, 2, 2)
plt.boxplot(df['Curah_Hujan'])
plt.title('Boxplot Curah Hujan')

plt.tight_layout()
plt.show()

#Visualisasi Dengan Scatterplot
plt.figure(figsize=(8, 4))
plt.scatter(df['Kelembapan_Udara'], df['Curah_Hujan'])
plt.xlabel('Kelembapan Udara')
plt.ylabel('Curah Hujan')
plt.title('Scatterplot Kelembapan Udara vs Curah Hujan')
plt.show()
```

```python
#Visualisasikan Standar Deviasi pada data Curah Hujan dan Suhu menggunakan Histogram

#Plot Histogram Untuk Curah Hujan
plt.figure(figsize=(8, 4))
plt.hist(df['Curah_Hujan'], bins=5, edgecolor='black')
plt.xlabel('Curah Hujan')
plt.ylabel('Frekuensi')
plt.title('Histogram Curah Hujan')

#Tambahkan Garis Vertikal Untuk Menunjukkan Nilai Mean
mean_curah_hujan = df['Curah_Hujan'].mean()
plt.axvline(mean_curah_hujan, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.text(mean_curah_hujan+1, 15, f'Mean: {mean_curah_hujan:.2f}', color='red')

#Tambahkan Garis Vertikal Untuk Menunjukkan Nilai Standar Deviasi
std_curah_hujan = df['Curah_Hujan'].std()
plt.axvline(mean_curah_hujan+std_curah_hujan, color='green', linestyle='dashed', linewidth=1, label='Std Dev')
plt.axvline(mean_curah_hujan-std_curah_hujan, color='green', linestyle='dashed', linewidth=1)
plt.text(mean_curah_hujan+std_curah_hujan+1, 15, f'Std Dev: {std_curah_hujan:.2f}', color='green')

#Plot Histogram Untuk Suhu
plt.figure(figsize=(8, 4))
plt.hist(df['Suhu_Udara'], bins=4, edgecolor='black')
plt.xlabel('Suhu')
plt.ylabel('Frekuensi')
plt.title('Histogram Suhu')

#Tambahkan Garis Vertikal Untuk Menunjukkan Nilai Mean
mean_suhu = df['Suhu_Udara'].mean()
plt.axvline(mean_suhu, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.text(mean_suhu+0, 15, f'Mean: {mean_suhu:.2f}', color='red')

#Tambahkan Garis Vertikal Untuk Menunjukkan Nilai Standar Deviasi
std_suhu = df['Suhu_Udara'].std()
plt.axvline(mean_suhu+std_suhu, color='green', linestyle='dashed', linewidth=1, label='Std Dev')
plt.axvline(mean_suhu-std_suhu, color='green', linestyle='dashed', linewidth=1)
plt.text(mean_suhu+std_suhu+0, 15, f'Std Dev: {std_suhu:.2f}', color='green')


plt.legend()
plt.show()
```

```python
#identifikasi Pola dan Anomali Data

#Menampilkan Informasi Umum Tentang Datast
print("informasi Umum:")
print(df.info())
print()

#Melihat Ringkasan Statistik Tentang Dataset
print("Ringkasan Statistik:")
print(df.describe())
print()

#Menghitung Korelasi antara Variabel Numerik
print("Korelasi:")
correlation = df.corr()
print(correlation)
print()

#Menemukan Anomali menggunakan Metode IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
outliers = ((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).sum()
print("Jumlah Anomali")
print(outliers)
print()
```
